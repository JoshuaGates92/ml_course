# Matplotlib Plotting and Data Visualisation 

## Anatomy of a Matplotlib Figure

### Workflow Example

```python
# 0. imports
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [ 11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customise plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")
plt.show()
# 5. Save and show:
# fig.savefig("filepath")
```

## Making figures with Numpy Arrays

```python
import numpy as np
```

```python
# Create data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and LINE PLOT
fig, ax = plt.subplots() #create figure and axes
ax.plot(x, x**2)
plt.show()
plt.close()
```

```python
# Use same data for scatter plots
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
plt.show()
plt.close()
```

```python
# another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x**0.5))
plt.show()
plt.close()
```

```python
# Make a plot from a dictionary
nut_butter_prices = {"Almond": 10,
                     "Cashew": 8,
                     "Peanut": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter Prices",
       xlabel="Butter Type",
       ylabel="Price ($)")
plt.show()
plt.close()
```

```python
# Horizontal bars
nut_butter_prices = {"Almond": 10,
                     "Cashew": 8,
                     "Peanut": 12}
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="Nut Butter Prices",
       ylabel="Butter Type",
       xlabel="Price ($)")
plt.show()
plt.close()
```

```python
# Histograms
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
ax.set(title="Histogram of 1000 Random Points in a Normal Distribution",
       xlabel="Standard Deviation",
       ylabel="Number of Observations")
plt.show()
plt.close()
```

```python
# The power of subplots (multiple plots in the same figure)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 6))
# Top left graph
ax1.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax1.set(title="Nut Butter Prices",
       xlabel="Butter Type",
       ylabel="Price ($)")

# Top right graph
ax2.hist(x)
ax2.set(title="Normal Distribution",
       xlabel="Standard Deviation",
       ylabel="Number of Observations")

# Bottom left
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax3.set(title="Nut Butter Prices",
       xlabel="Butter Type",
       ylabel="Price ($)")

# Bottom right graph
ax4.hist(x)
ax4.set(title="Normal Distribution",
       xlabel="Standard Deviation",
       ylabel="Number of Observations")


plt.show()
plt.close()
# Formatting needs some work, but we get it
```

```python
# Remember regex fun from pandas section? I do. Get rid of the $ and decimal point in the Price column
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]', '')
car_sales['Price'] = car_sales['Price'].str[:-2]

# Did it work?
car_sales
```

<font color="green">Yes it did! Look at me bringing in code from other sections like I know what I am doing!</font>

```python
car_sales['Sales Date'] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
# Add total sales ($) column
car_sales['Total Sales ($)'] = car_sales['Price'].cumsum()
car_sales
```

Uh oh, forgot our Price column is actually a string so it's just concatenating the strings... cool trick though! Let's fix it.

```python
car_sales['Total Sales ($)'] = car_sales['Price'].astype(int).cumsum()
car_sales
```

```python
# Time to plot total sales
car_sales.plot("Sales Date", "Total Sales ($)")
plt.show()
plt.close()
```
